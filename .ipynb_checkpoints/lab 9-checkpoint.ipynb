```python
# 관련 패키지 import
import selenium
from selenium import webdriver
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By 
import time
```

```python
# 1. webdriver 객체 생성
driver = webdriver.Chrome('./driver/chromedriver')
```

```python
url = 'https://www.megabox.co.kr/movie-detail/comment?rpstMovieNo=21049700'
driver.get(url)
```

```python
sel = '#contentData > div > div.movie-idv-story > ul > li:nth-child(2) > div.story-area > div.story-box > div > div.story-cont'
```

```python
elem = driver.find_elements(By.CSS_SELECTOR,sel)
```

```python
sel_point = '#contentData > div > div.movie-idv-story > ul > li:nth-child(2) > div.story-area > div.story-box > div > div.story-cont > div.story-point'
sel_rec = '#contentData > div > div.movie-idv-story > ul > li:nth-child(2) > div.story-area > div.story-box > div > div.story-cont > div.story-recommend'
sel_txt= '#contentData > div > div.movie-idv-story > ul > li:nth-child(2) > div.story-area > div.story-box > div > div.story-cont > div.story-txt'
elem_point = driver.find_elements(By.CSS_SELECTOR,sel_point)
elem_rec = driver.find_elements(By.CSS_SELECTOR,sel_rec)
elem_txt = driver.find_elements(By.CSS_SELECTOR,sel_txt)
```

```python
elem_point
```

```python
result = []
```

```python
for e in elem_point :
    res_p = e.text.split('\n')
    #print(res)
for e in elem_rec :
    res_r = e.text.split('\n')
    #print(res)
for e in elem_txt :
    res_t = e.text.split('\n')
    #print(res)
result.append(pd.DataFrame({'평점':res_p[0],'추천':res_r[0],'한줄평':res_t[0]},index=range(0,1)))
```

```python
result = []
for i in range(1, 11):
    sel_page = f'#contentData > div > div.movie-idv-story > nav > a:nth-child({i})'
    for j in range(2, 12):
        sel_point = f'#contentData > div > div.movie-idv-story > ul > li:nth-child({j}) > div.story-area > div.story-box > div > div.story-cont > div.story-point'
        sel_rec = f'#contentData > div > div.movie-idv-story > ul > li:nth-child({j}) > div.story-area > div.story-box > div > div.story-cont > div.story-recommend'
        sel_txt = f'#contentData > div > div.movie-idv-story > ul > li:nth-child({j}) > div.story-area > div.story-box > div > div.story-cont > div.story-txt'
        res_p = [e.text.split('\n')[0] for e in driver.find_elements(By.CSS_SELECTOR, sel_point)]
        res_r = [e.text.split('\n')[0] for e in driver.find_elements(By.CSS_SELECTOR, sel_rec)]
        res_t = [e.text.split('\n')[0] for e in driver.find_elements(By.CSS_SELECTOR, sel_txt)]
        result.append(pd.DataFrame({'평점': res_p, '추천': res_r, '한줄평': res_t}))
```

```python
result
```

```python
df = pd.concat(result).reset_index(drop=True)
df
```

```python
driver.close()
```

```python
pd.concat(result).reset_index(drop=True).to_csv('./data/movie.csv')
```
