```python
result = []
for i in range (1, 11) :
    sel_page = '#contentData > div > div.movie-idv-story > nav > a:nth-child('+str(i)+')'
    for j in range(2, 12) : 
        sel_point = '#contentData > div > div.movie-idv-story > ul > li:nth-child('+str(j)+') > div.story-area > div.story-box > div > div.story-cont > div.story-point'
        sel_rec = '#contentData > div > div.movie-idv-story > ul > li:nth-child('+str(j)+') > div.story-area > div.story-box > div > div.story-cont > div.story-recommend'
        sel_txt= '#contentData > div > div.movie-idv-story > ul > li:nth-child('+str(j)+') > div.story-area > div.story-box > div > div.story-cont > div.story-txt'
        elem_point = driver.find_elements(By.CSS_SELECTOR,sel_point)
        elem_rec = driver.find_elements(By.CSS_SELECTOR,sel_rec)
        elem_txt = driver.find_elements(By.CSS_SELECTOR,sel_txt)
        for e in elem_point :
            res_p = e.text.split('\n')
            #print(res)
        for e in elem_rec :
            res_r = e.text.split('\n')
            #print(res)
        for e in elem_txt :
            res_t = e.text.split('\n')
            #print(res)
        result.append(pd.DataFrame({'평점':res_p[0],'추천':res_r[0],'한줄평':res_t[0]},index=range(0,1)))
```
